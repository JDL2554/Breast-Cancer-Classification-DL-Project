```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#uploading data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading it into a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

**Building the Actual Neural Network**

pj.jpg

```python
#importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the Neural Network Layers
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

dog- 0\

```python
#training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

**Visualizing Accuracy and Loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['train data', 'validation data'], loc='lower right')
#first epoch is labled starting at 0 instead of 1 and then they go to epoch 9 instead of 10
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['train data', 'validation data'], loc='upper right')
#first epoch is labled starting at 0 instead of 1 and then they go to epoch 9 instead of 10
```

**Accuracy of the Model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

**Building the Predictive System**

```python
input_data = (20.57, 17.77, 132.9, 1326.0, 0.08474, 0.07684, 0.0869, 0.07017, 0.1812, 0.05667, 0.5435, 0.7739, 3.398, 74.08, 0.005225, 0.01308, 0.0186, 0.0134, 0.01389, 0.003532, 24.99, 23.41, 158.8, 1956, 0.1238, 0.1866, 0.2416, 0.186, 0.275, 0.08902)

#change the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the np array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardizing the input_data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The Tumor is Malignant')
elif(prediction_label[0] == 1):
  print('The Tumor is Benign')
```
